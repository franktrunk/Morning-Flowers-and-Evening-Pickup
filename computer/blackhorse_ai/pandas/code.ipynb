# 1.Series 对象的创建


## 1.1 numpy的ndarray => series对象

```python
# 导包
import numpy as np
import pandas as pd
```

```python
# 1. 创建ndarray对象
arr = np.array([1,2,3,4,5])
print(arr)
print(type(arr))
```

```python
s1 = pd.Series(data = arr)
print(s1)
print(type(s1))
```

## 直接传入python列表，构建series对象

```python
# 没有指定索引值，默认是0-n
# s2 = pd.series(data = ['乔峰','男',33])
s2 = pd.Series( ['乔峰','男',33]) # data参数名可以省略不写
print(s2)
print(type(s2))
```

## 1.3 传入python列表，构建series对象，并指定行索引

```python
s3 = pd.Series( ['乔峰','男',33],index = ['name','gender','age']) # data参数名可以省略不写
print(s3)
print(type(s3))
```

## 1.4 通过元组的方式创建series对象

```python
s4 = pd.Series( ('乔峰','男',33),index = ['name','gender','age']) # data参数名可以省略不写
print(s4)
print(type(s4))
```

## 1.5通过字典的方式创建

```python
s5 = pd.Series({'name':'乔峰','gender':'男','age':33})
print(s5)
print(type(s5))
```

# 2.创建DataFrame对象

## 2.1字典方式构建df对象

```python
# 创建字典，记录数据
dict_data = {
    'id': [1,2,3],
    'name':['乔峰','虚竹','段誉'],
    'age':[33,29,21]
}
# 基于上述字典构建
df1 = pd.DataFrame(dict_data,index=['A','B','C'])# 指定行索引
df1 = pd.DataFrame(dict_data,index=['A','B','C'],columns=['id','age'])# 指定列名
print(df1)
print(type(df1))
```

## 2.2 列表+元组构建df

```python
# 列表+元组，一个元组 = 一行数据
list_data = [(1,'乔峰',33),(2,'虚竹',29),(3,'段誉',21)]
# 构建df对象
# index：指定行索引 ，columns：指定列名（的顺序）
df2 = pd.DataFrame(list_data,index =['x','Y','Z'],columns=['id','name','age'])
# 打印
print(df2)
df2 = pd.DataFrame(list_data,index =['x','Y','Z'],columns=['id','age','name'])
print(df2)
print(type(df2))
```

# 4.series对象的常用方法介绍

```python
s1 = pd.Series([1,2,3,4,2,3],index=['A','B','C','D','E','F'])
print(s1)
```

```python
# 常用法方法
print(len(s1)) #长度
print(s1.size) # 元素个数
print(s1.head(n=2))# 头（默认五个）
print(s1.tail())#尾 （默认五个）
print(s1.tail(n=3))

print(s1.keys()) # 索引
print(s1.index) # 索引

print(s1.tolist()) # 转列表
print(type(s1.tolist())) 

print(s1.to_frame()) # 转df对象
print(type(s1.to_frame()))
```

```python
print(s1.describe()) # 查看详细信息
print(s1.max())
print(s1.min())
print(s1.mean())
print(s1.std())
print(s1.drop_duplicates())# 去重，返回series对象
print(s1.unique()) # 去重，返回数组
```

```python
print(s1.sort_values()) # 根据值进行排序，默认升序
print(s1.sort_values(ascending= False)) # 降序

print(s1.sort_index()) # 根据索引进行排序，默认升序
print(s1.sort_index(ascending= False)) # 降序

```

```python
print(s1.value_counts()) # 统计每个值出现的次数
```

# 6.serie的布尔值操作

```python
# 读取整数数据，获取df对象
df = pd.read_csv('scientists.csv')
df
```

```python
# 手动出纳入布尔值
bool_list = [True,False,False,True,False,False,True,True]
df[bool_list]
```

```python
# 根据条件传入bool值，筛选数值
# 要求：筛选出年龄大于平均年龄的科学家
age_series = df['Age']
age_series
```

```python
avg_age = age_series.mean()
avg_age
```

```python
# 判断条件，生成bool列表
age_series > avg_age
```

```python
age_series[age_series>avg_age]
```

```python
# 简化版本
print(df['Age'][df['Age']> df['Age'].mean()]) # 只获取符合条件的元素
print(df[df['Age']> df['Age'].mean()]) # 整行数据都拿出来了
```

# 7. series的运算

## 7.1 series和数值运算，数值会和series的每个值运算

```python
print(age_series+10)
print(age_series*2)
```

## 7.2 两个series之间的计算，如果长度一致，会按照对应的元素进行逐个计算

```python
age_series+age_series
```

## 7.3 两个series之间计算，如果长度不一致，则对应元素计算，不匹配的元素用nan填充

```python
age_series + pd.Series([1,10])
```

##  7.4 series之间进行计算时，数据尽可能会依据索引来计算，优先计算索引一样的数据

```python
#  对原数据，按照年龄降序排列，获取新的series对象
rev_series = age_series.sort_values(ascending= False)
rev_series
```

```python
age_series # 原始对象
```

```python
age_series+rev_series
```
